## Pilha e Fila


### Pilha

A pilha é uma estrutura comumente ultilizado na programação, ela segue e se baseia em uma lista que respeita esquema FILO (*First in, First Out*), Ou seja, o valor inserido sempre será colado na ultima posição, e o valor retirado sempre sairá da ultima posição também. Como se fosse uma pilha de livros onde só pode ser pego e tirados livros por cima.

Sendo o código base esse:

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items) - 1]

    def is_empty(self):
        return self.items == []

    def size(self):
        return len(self.items)
```

Onde, 

*   A função **Stack** cria uma pilha vazia, representada no codigo a cima pelo __init__, já que o __init__ puxa o nome da classe.

*   **Push(item)** adiciona um valor na primeira posição da pilha.

*   A função **Pop** retira o valor que esta alocado na primeira posição.

*   **Peek** mostra o valor que está na primeira posição.

*   **is_empty** verifica se a pilha está vazia.

*   **size** mostra quantos valores tem na pilha.

estes são as funções bases que caracterizam uma pilha, porém podem ser adicionadas outras funções para se adequar melhor em cada aplicação.

### Fila


Fila é uma estrutura de dados, a qual, se baseia em uma lista que respeita o conceito de FIFO (*first in, first out*). Ou seja, na hora de inserir elementos/valores os mesmos sempre serão colocados na primeira posição, e na hora da saíde de valores sempre será retirado o primeiro valor colado em ordem de chegada. Bem representado por uma fila convencional de pessoas.

Sendo o código base:

```python
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0, item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)
```

Onde,

*   A função **Queue** cria uma fila vazia, representada no codigo a cima pelo __init__, já que o __init__ puxa o nome da classe.

*  **enqueue(item)** ele inseri um novo dado de acordo com a ordem de chegada.

*  **dequeue()** tira o primeiro em questão de chegada.

*  **is_empty** testa se a fila ta vazia.

*  **size()** retorna o numero de itens da fila.

Assim como a pilha, essas são as funções bases que caracterizam a fila, porém podem ser adicionadas outras funções para se adequar melhor em cada aplicação.

## Pesquisas


### Pesquisa Sequencial

Basicamente esse tipo de pesquisa vai buscar o valor de interesse de forma linear. Então ele vai partir de um ponto (normalmente o valor inicial da lista, index 0) vai perconcorrer todos os valores, só parando se encontrar o valor de desejo ou encontrar um valor maior que o valor de desejo ou ainda acabar o campo de busca (a lista).

```python
#Sequencial Padrão
def sequential_search(a_list, item):
    pos = 0
    found = False
    while pos < len(a_list) and not found:
        if a_list[pos] == item:
            found = True
        else:
             pos = pos + 1

    return found


test_list = [1, 2, 32, 8, 17, 19, 42, 13, 0]
print(sequential_search(test_list, 3))
print(sequential_search(test_list, 13))


#Sequencial Ordenada, ele tem o "stop" a mais, o "stop" se já passou do valor desejado, o valor que se esta buscando.
def ordered_sequential_search(a_list, item):
    pos = 0
    found = False
    stop = False
    while pos < len(a_list) and not found and not stop:
        if a_list[pos] == item:
            found = True
        else:
            if a_list[pos] > item:
                stop = True
            else:
                pos = pos + 1

    return found


#test_list = [0, 1, 2, 8, 13, 17, 19, 32, 42,]
#print(ordered_sequential_search(test_list, 3))
#print(ordered_sequential_search(test_list, 13))
```

### Pesquisa Binaria

Diferente da busca sequencial essa busca separa o campo de busca (a lista) em duas partes, assim ele identifica se o valor que deseja está a cima ou abaixo do midpoint, que seria onde foi divido, dessa forma identifica onde esta o valor desejado, e depois repete o processo até encontrar o valor desejado ou perceber que o valor desejado não está no campo de busca.

```python
def binary_search(a_list, item):
    first = 0
    last = len(a_list) - 1
    found = False

    while first <= last and not found:
        midpoint = (first + last) // 2
        if a_list[midpoint] == item:
            found = True
        else:
            if item < a_list[midpoint]:
                last = midpoint - 1
            else:
                first = midpoint + 1
        print(a_list[midpoint], midpoint)

    return found


test_list = [0, 1, 2, 8, 13, 17, 19, 32, 42]
print(binary_search(test_list, 19)) 
```

Logo a baixo o codigo de pesquisa binaria forma recursiva. Recursividade basicamente chama novamente a função dentro da propria função, esse "loop" só é interompido quando uma condição é respeitada.

```python
def binary_search(a_list, item):
    if len(a_list) == 0:
        return False
    else:
        midpoint = len(a_list) // 2
        
    if a_list[midpoint] == item:
        return True
    else:
        if item < a_list[midpoint]:
            return binary_search(a_list[:midpoint], item)
        else:
            return binary_search(a_list[midpoint + 1:], item)


test_list = [0, 1, 2, 8, 13, 17, 19, 32, 42]
print(binary_search(test_list, 3))
print(binary_search(test_list, 2))
print(binary_search(test_list, 52))
print(binary_search(test_list, -1))
# Uma operação de recursividade tende a ser igual ou pior a uma operação dinamica
```

## Ordenação


### Bubble Sort

Compara os valores de um em um. Então ele vai na primeira posição e compara com todos na lista, e em seguida irá na segunda posição e comparará com todos na lista, assim por diante.


```python
def bubble_sort(a_list):
    for pass_num in range(len(a_list) - 1, 0, -1):
        print("Estou no primeiro for")
        for i in range(pass_num):
            print("Estou no segundo for")
            if a_list[i] > a_list[i + 1]:
                temp = a_list[i]
                a_list[i] = a_list[i + 1]
                a_list[i + 1] = temp


#a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
a_list = [1, 2, 3, 4, 5, 6]
bubble_sort(a_list)
print(a_list)
```

### Ordenação por Inserção

Cria uma sublista na propria lista, dessa forma o algoritmo compara os valores da lista com os valores da sublista, assim a sublista inicia com um valor e vai se organizando a partir das comparações e no final a sublista se torna a lista, e assim consegue uma lista ordenada.

### Ordenação MargeSort

Marge sorte segue o conceito de dividir para conquistar, então ele vai dividir todos os itens da lista e depois ir os juntando já os organizando, muito bem representado pelas imagens a baixo:

Sendo o código:

```python
def merge_sort(a_list):
    print("Splitting ", a_list)
    if len(a_list) > 1:
        mid = len(a_list) // 2
        left_half = a_list[:mid]
        right_half = a_list[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = 0
        j = 0
        k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                print("valor esquerdo menor que valor direito")
                a_list[k] = left_half[i]
                i = i + 1
            else:
                print("valor direito menor que valor esquerdo")
                a_list[k] = right_half[j]
                j = j + 1
            k = k + 1

        while i < len(left_half):
            print("Lado esquerdo")
            a_list[k] = left_half[i]
            i = i + 1
            k = k + 1

        while j < len(right_half):
            print("Lado direito")
            a_list[k] = right_half[j]
            j = j + 1
            k = k + 1
    print("Merging ", a_list)


a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
merge_sort(a_list)
print("Final: ", a_list)
```

### Ordenação QuickSort

O QuickSort segue o conceito de dividir para conquistar, o algoritmo tem a ideia de determinar o pivo e posteriormente colocar ele como divisor de duas partes uma menor que o pivo outra maior que o pivo.

```python
def quick_sort(a_list):
    quick_sort_helper(a_list, 0, len(a_list) - 1)


def quick_sort_helper(a_list, first, last):
    if first < last:
        split_point = partition(a_list, first, last)
        quick_sort_helper(a_list, first, split_point - 1)
        quick_sort_helper(a_list, split_point + 1, last)


def partition(a_list, first, last):
    print("Partition", a_list)
    pivot_value = a_list[first]

    left_mark = first + 1
    right_mark = last
    done = False

    while not done:
        while left_mark <= right_mark and a_list[left_mark] <= pivot_value:
            left_mark = left_mark + 1

        while a_list[right_mark] >= pivot_value and right_mark >= left_mark:
            right_mark = right_mark - 1

        if right_mark < left_mark:
            done = True
        else:
            temp = a_list[left_mark]
            a_list[left_mark] = a_list[right_mark]
            a_list[right_mark] = temp

    temp = a_list[first]
    a_list[first] = a_list[right_mark]
    a_list[right_mark] = temp

    return right_mark


#a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
#a_list = ["a", "b", "d", "z", "c"]
#a_list = ["a", "A", "b", "d", "C"]
a_list = ["cesar", "abacaxi", "joelho", "café"]
quick_sort(a_list)
print(a_list)
```
